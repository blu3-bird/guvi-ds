##### f-string exercises

```python
user_name = 'arson'
age = 34

print(f'username is {user_name} and age is {age}') # f-string (syntax is print(f'text {var}'))

print(f'heyyy {54+45}') # not only variables, we can also do other things like math operations.
print(f'checking dtypes: {type(age)}') # we can even use different func in f-string.
```

##### string functions

```python
name = 'Portgas d. Arson'
print('Practing Str functions:')
print(f'Lower case: {name.lower()}') # to convert all the letters to lower case
print(f'Upper case: {name.upper()}') # to all letters to upper case
print(f'swapcase: {name.swapcase()}') # it will swap the letter case
print(f'Captailize: {name.capitalize()}') # captialize
print(f'Title : {name.title()}') # title
```

```python
# removing white spaces from string
xyz = "   arson d   " 
print(f'left strip: {xyz.rstrip()}') # removed the white space from the right side
print(f'right strip: {xyz.lstrip()}') # removed the white space from the left side
print(f'Strip both: {xyz.strip()}') # removed the white space from the both side
```

##### alpha, alpha num and num func

```python
data = "laptop638"
udfa = "aldfd jdf" # adding space in var
print(f'Alpha: {data.isalpha()}') # Only alpha letters
print(f'Alpha Num: {data.isalnum()}') # either alpha or digits
print(f'Digit: {data.isdigit()}') # only digit

print(f'\nAdded space in var value: {udfa.isalnum()}') # if we have any space or special characters we ll get output false, coz it will treat it as special characters. not as str or digit
```
