##### f-string exercises

```python
user_name = "arson"
age = 34

print(
    f"username is {user_name} and age is {age}"
)  # f-string (syntax is print(f'text {var}'))

print(
    f"heyyy {54 + 45}"
)  # not only variables, we can also do other things like math operations.
print(f"checking dtypes: {type(age)}")  # we can even use different func in f-string.
```

##### string functions

```python
name = "Portgas d. Arson"
print("Practing Str functions:")
print(f"Lower case: {name.lower()}")  # to convert all the letters to lower case
print(f"Upper case: {name.upper()}")  # to all letters to upper case
print(f"swapcase: {name.swapcase()}")  # it will swap the letter case
print(f"Captailize: {name.capitalize()}")  # captialize
print(f"Title : {name.title()}")  # title
```

```python
# removing white spaces from string
xyz = "   arson d   "
print(f"left strip: {xyz.rstrip()}")  # removed the white space from the right side
print(f"right strip: {xyz.lstrip()}")  # removed the white space from the left side
print(f"Strip both: {xyz.strip()}")  # removed the white space from the both side
```

##### alpha, alpha num and num func

```python
data = "laptop638"
udfa = "aldfd jdf"  # adding space in var
print(f"Alpha: {data.isalpha()}")  # Only alpha letters
print(f"Alpha Num: {data.isalnum()}")  # either alpha or digits
print(f"Digit: {data.isdigit()}")  # only digit

print(
    f"\nAdded space in var value: {udfa.isalnum()}"
)  # if we have any space or special characters we ll get output false, coz it will treat it as special characters. not as str or digit
```

##### List

```python
is_married = False
data = [
    "Arson",
    "laptop",
    "chair",
    "phone",
    89,
    is_married,
]  # list is hetro, means we can save str,bool,int
print(f"Data types of list: {type(data)}\n")
print(f"Printing the list elements:\n{data}\n")

# addinng new element in list
data.append("book")

print(f"update list(added new element:\n{data})\n")

# adding new multiple elements in exisiting list
# extend() func

new_data = ["guitar", "charger", "enough"]
data.extend(new_data)  # syntax: list.extend(new_list)

print(f"updated list(added multiple elements):\n{data}\n")

# adding new elements using extend but a str not list
new_data.extend("yoyo")
print(f"updated list:(added new elements instead of list): {new_data}")
```

##### Slicing

```python
seven = [4, 5, 6, 8, 90]
print(f"callign different elements of a list: {seven[1:3]}")
print(f"callinng index from right to left: {seven[-4:-2]}")

print(
    f"callinng index from right to left: {seven[-2:-4]}"
)  # getting no elements for this one, cause we started -2 index which is 8 and end at -4 which is 5, so there are so elements between this range, so it returned empty list.
```

```python
# insert func()

gfd = []
gfd.extend("ABCCCDEF")
gfd.insert(2, "arson")  # list.insert(index, value)
print(f"updated list after using insert func: {gfd}")

# pop func
gfd.pop()  # deletes the last elements
print(
    f"updating list using pop func and index: {gfd.pop(2)}"
)  # remove the index elements


# remove func
gfd.remove("C")  # syntax: list.remove(element)
print(
    f"updated list after remove func(): {gfd}"
)  # it will only remove the first duplicates not all the matching elements
```

##### Set

```python
book = set()  # unordered #we can change/update

book.add("book1")
book.add("book2")
book.add("book2")
print(
    f"printing set: {book}"
)  # duplicates are not allowed, keeeps only first duplicates

# to add multiple elements in set: update func
# update func()

book.update(
    "5", ["book5"], ["novel"]
)  # keeping book5 in sq brackets means we are treating it as a list. and book5 is its elements.
# if we add a str like 'arson', it will treat arson as a list and its char as list elements
print(f"updated set after update func(): {book}")
```

##### tuple

```python
eight = ("a", "tij", "ok", "bien")
print(f"printing tuple: {eight}")

print(f"calling element is tuple: {eight[2]}")
```

```python
# dictionary

wallet = {"money": 1000, "has_money": True}
print(f"printing dictionary: {wallet}\n")

# adding new item in dict
wallet["proofs"] = ["id-card", "credit-card", "DL"]
print(f"updated dict: {wallet}\n\n")

if wallet["money"] > 0:
    wallet["has_money"] = True
else:
    wallet["has_money"] = False

print(f"updated dict:{wallet}\n")

# as earlier i have added a list in one item, so we can access elemnts of that list using indexing
print(f"accessing elements of a list inside a dict: {wallet['proofs'][1]}")
```
