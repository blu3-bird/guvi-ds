```python
scores = [12,14,15,18,20,12]
max = 20
percent = []

for i in scores:
    value = (i*100)/max
    percent.append(value)
print(percent)


```

```python
# while looop
scores = [12,14,15,18,20,12]
max_score = 20
i = 0
percent = []

while i < len(scores):
    values = (scores[i] * 100)/max_score
    percent.append(values)
    i = i+1
print(percent)
```
