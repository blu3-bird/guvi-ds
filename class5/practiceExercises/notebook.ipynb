#### Exercise 1.3: String Analysis
```python
# Write a program that:
# 1. Counts vowels and consonants in a string
# 2. Finds the first occurrence of each character
# 3. Checks if two strings are anagrams
# 4. Reverses a string without slicing
```

````python
def string_analysis(text1):
    vowel_count = 0
    consonants_count = 0
    vowels = 'aeiouAEIOU'

    for char in text1:
        if char in vowels:
            vowel_count +=1
        else:
            consonants_count +=1
    
    firstOccurrance = {} # defining a dict to store the index of each char.
    for idx,char in enumerate(text1): 
        if char not in firstOccurrance: # if char is not present in dict, it means it is his first occurance,
            firstOccurrance[char] = idx
    

    reverse_text = ""
    for char in text1:
        reverse_text = char + reverse_text # S[1] + S[0] and so onnn.
    
    return{
        'vowel_count' : vowel_count,
        'consonants_count' : consonants_count,
        'reverse_text' : reverse_text,
        'first_occurrance' : firstOccurrance
    }
S = input()
string_analysis(S)
````

````python
def is_prime(num:int) -> bool:
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    else:
        for i in range(3,int(num**0.5),2):
            if num % i == 0:
                return False
        return True

is_prime(5)
````

````python
def gcd(a,b:int) -> int:
    """to find largest int that divides both numbers"""
    results_a = []
    results_b = []
    for i in range(1,a+1):
        if a % i == 0:
            results_a.append(i)
    
    for i in range(1,b+1):
        if b % i == 0:
            results_b.append(i)
    results = []
    for i in results_a:
        if i in results_b:
            results.append(i)
    return max(results)

gcd(6,8)
````

````python
def lcm(a,b:int) -> int:
    results_a = []
    results_b = []

    for i in range(1,100):
        a_v = a * i
        b_v = b * i
        results_a.append(a_v)
        results_b.append(b_v)

        for num in results_a:
            if num in results_b:
                return num
                


lcm(12,18)
````

````python
def power(base:int, exp:int) -> int:
    """Calculate base^exp"""
    if base <= 0: return 0
    results = 1 # defining results value, (in case exp = 0)
    for _ in range(exp): # to run our loop exp numbers of times
        results = results * base 
    return results

power(2,3)
````

````python
def sqrt_approx(n:int) -> int:
    """Approximate square root using binary search"""
    if n < 0: raise ValueError('Int is less than zero')
    if n == 0 or n == 1: return 1

    low, high = 1,n # (sqrt must be between 1 and n)
    while low <= high:
        mid = (low + high) // 2
        sq = mid * mid

        if sq == n:
            return mid
        elif sq < n:
            low = mid + 1
        else:
            high = mid - 1
    return high
sqrt_approx(1)
````

````python
def factorial(n:int) -> int:
    """Calculate factorial recursively and iteratively"""
    value = 1
    for i in range(1,n+1):
        value = value * i
    return value

factorial(4)
````

````python
def reverse_list(lst:list) -> list:
    """Return reversed list"""
    return lst[::-1]
arr = [1,8,7,5]
reverse_list(arr)
    
````

````python
def remove_duplicates(lst:list)-> list:
    """Remove duplicates preserving order"""
    duplicates = []
    for i in lst:
        if i in duplicates:
            lst.remove(i)
        else:
            duplicates.append(i)
    return duplicates
arr = [2,2,2,1,4,1]
remove_duplicates(arr)
    
````

````python
def merge_sorted_lists(list1:list, list2:list) -> list:
    """Merge two sorted lists"""
    results = []
    for i in list1:
        results.append(i)
    
    for i in list2:
        results.append(i)
    return results
lst = [1, 3 , 5]
lst2 = [2, 4]
merge_sorted_lists(lst,lst2)
````

````python
# Problem 1: Product of Digits
# Given N, print product of its digits
# Input: 2143 → Output: 24
N = input()
product = 1
for char in N:
    product *= int(char)
print(product)
````

````python
# Problem 2: String Middle Modification
# Replace middle character(s) with '*'
# Input: "hello" → Output: "he*lo"
# Input: "test" → Output: "te**"
S = input().strip()
mid = len(S) // 2
if len(S) % 2 != 0: 
    print(f'{S[:mid]}*{S[mid+1:]}')
else:
    print(f'{S[:mid]}**{S[mid+2:]}')
````

````python
# Problem 3: Composite Number Check
# Print 'yes' if composite, else 'no'
factors = []
def is_composite(num):
    for i in range(1,num+1):
        if num % i == 0:
            factors.append(i)
    
    if len(factors) > 2:
        return True
    

N = int(input())
if is_composite(N):
    print('yes')
else:
    print('noo')
````

````python
# Problem 4: Factors of a Number
# Print all factors of N

def factors(num):
    num_factors = []
    for i in range(1,num+1):
        if num % i == 0:
            num_factors.append(i)
        
    return num_factors
N = int(input())
factors(N)
````

````python
# Problem 5: Perfect Square Product
# Given N, M check if N*M is perfect square

def perfect_sq(a:int,b:int) -> bool:
    if a*b == a**2:
        return True
    else:
        return False
N,M = map(int,input().split())
perfect_sq(N,M)
````

````python
# Problem 6: Vowel Counter
# Count vowels in a string (case-insensitive)
def vowel_count(S):
    count = 0
    vowels = 'aeiou'

    for char in S:
        if char.lower() in vowels:
            count +=1

    return count
s = input()
vowel_count(s)
````

````python
# Problem 7: String Length Without len()
# Find length without using len() function
def str_len(S):
    count = 0
    for _ in range(len(S)):
        count +=1
    return count
s = input()
str_len(s)
````

````python
# Problem 8: Case-Sensitive Equality
# Compare two strings case-sensitively
case_s = lambda a,b : a.lower() == b.lower()
case_s('hello','ello')
````

````python
# Problem 9: Common Characters
# Check if two strings share any character
def shared_chars(a:str,b:str) -> bool:
    common_char = filter(lambda char: char in b,a)
    return bool(common_char)

shared_chars('helo','hello')

````

````python
def comman_s(a:str,b:str) -> bool:
    for char in a:
        if char in b:
            return True
        else:
            return False

comman_s('q','hello')
````

````python
# Problem 10: Round to Nearest Greater Integer
# Input: 2.6 → Output: 3
round_to = lambda x: int((x*100)//100) + 1
round_to(1.3)
````
