#### Exercise 1.3: String Analysis
```python
# Write a program that:
# 1. Counts vowels and consonants in a string
# 2. Finds the first occurrence of each character
# 3. Checks if two strings are anagrams
# 4. Reverses a string without slicing
```

````python
def string_analysis(text1):
    vowel_count = 0
    consonants_count = 0
    vowels = 'aeiouAEIOU'

    for char in text1:
        if char in vowels:
            vowel_count +=1
        else:
            consonants_count +=1
    
    firstOccurrance = {} # defining a dict to store the index of each char.
    for idx,char in enumerate(text1): 
        if char not in firstOccurrance: # if char is not present in dict, it means it is his first occurance,
            firstOccurrance[char] = idx
    

    reverse_text = ""
    for char in text1:
        reverse_text = char + reverse_text # S[1] + S[0] and so onnn.
    
    return{
        'vowel_count' : vowel_count,
        'consonants_count' : consonants_count,
        'reverse_text' : reverse_text,
        'first_occurrance' : firstOccurrance
    }
S = input()
string_analysis(S)
````

#### Exercise 2.2: Set Operations
```python
# Given two sets of student IDs:
class_a = {101, 102, 103, 104, 105}
class_b = {103, 104, 106, 107, 108}

# Find:
# 1. Students in both classes (intersection)
# 2. All unique students (union)
# 3. Students only in class_a (difference)
# 4. Students in either class but not both (symmetric difference)
# 5. Add a new student to class_a
# 6. Remove a student from class_b
```

````python
class_a = {101, 102, 103, 104, 105}
class_b = {103, 104, 106, 107, 108}

#1. Students in both classes (intersection)
# common_student = class_a.intersection(class_b)
common_student = class_b&class_a
print(f'common student: {common_student}')

# 2. All unique students (union)
# all_student = class_a.union(class_b)
all_student = class_a|class_b
print(f'All student: {all_student}')

# 3. Students only in class_a (difference)
# class_a_student = class_a.difference(class_b)
class_a_student = class_a - class_b
print(f'class A student {class_a_student}')

# 4. Students in either class but not both (symmetric difference)
# student_in_one_class = class_b.symmetric_difference(class_a)
student_in_one_class = class_a ^ class_b
print(f'student_in_one_class {student_in_one_class}')

# Add a new student to class_a
new_student = int(input('Add a new student to class_a'))
class_a.add(new_student)
print(f'class A after adding new student: {class_a}')
# 6. Remove a student from class_b
remove_student = int(input(' Remove a student from class_b'))
class_b.discard(remove_student) # using discard here, instead of remove, discard checks if elements exists, if not, it pass
print(f'Updated Class B : {class_b}')
````

#### Exercise 3.1: Number Classifier
```python
# Write a program that classifies a number as:
# 1. Positive, Negative, or Zero
# 2. Even or Odd
# 3. Prime or Composite
# 4. Perfect Square or Not
# 5. Armstrong Number or Not (e.g., 153 = 1³ + 5³ + 3³)
```

````python
N = int(input())

# 1. Positive, Negative, or Zero
def int_check(num):
    if num == 0:
        arson = 'Zero'
    elif num < 0:
        arson = '-Ve'
    else:
        arson = '+Ve'
    return arson
# 2. Even or Odd
def odd_even(num):
    if num % 2 == 0:
        arson = 'Even'
    else:
        arson = 'Odd'
    return arson

# 3. Prime or Composite
def chk_prime(num):
    if num <=1:
        return 'Not Prime'
    elif num == 2:
       return 'Prime'
    elif num % 2 == 0:
        return 'Not Prime'
    else:
        for ch in range(3,num,2):
            if num % ch == 0:
                return 'Not Prime'
        return 'Prime'
        # return arson


# 4. Perfect Square or Not
def perfect_sq(num):
    root = num ** 0.5

    if root == int(root):
        arson = 'Perfect sq'
    else:
        arson = 'imperfect sq'
    return arson
# 5. Armstrong Number or Not (e.g., 153 = 1³ + 5³ + 3³)
def armstrong_num(num):
    value = []
    arson = 'no'
    str_num = str(num)
    for i in (str_num):
        chk = int(i) ** len(str_num)
        value.append(chk)
    if sum(value) == num:
        arson = 'yes'
    return arson
print(int_check(N))
print(odd_even(N))
print(chk_prime(N))
print(perfect_sq(N))
print(armstrong_num(N))

````

### **Section 4: Loops (For & While)**

#### Exercise 4.1: Pattern Printing
```python
# Print these patterns using loops:
# 1. Right triangle of stars
# 2. Pyramid pattern
# 3. Number triangle (1, 12, 123, 1234)
# 4. Floyd's triangle
```

````python
# Print these patterns using loops:
# 1. Right triangle of stars
N = int(input())
i = 1
while i < N+1:
    print('*'*i)
    i = i+1

# 2. Pyramid pattern
for i in range(1,N+1):
    row = " "*(N-i) +'*'*(2*i-1)
    print(row)
# 3. Number triangle (1, 12, 123, 1234)
go = ''
for idx,n in enumerate(range(0,N)):
    go = f'{go}{n+1}'
    print(go)

# 4. Floyd's triangle
num = 1
for row in range(1,N+1):
    for _ in range(row):
        print(num,end='')
        num += 1
    print()
````
