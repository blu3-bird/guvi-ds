```python
import math

X = 4.9
print(math.floor(X)) # return the int part of the float
print(math.ceil(X)) # round off to the next int. 
print(math.sqrt(50)) # sqrt root.
print(math.isqrt(50)) # when we want int as output.


```

#### Functions


```python
def greet():
    print('welcome')
def bye():
    print('bye')

def display(fun):
    fun()

display(bye)
```

```python
x= 10
y = 3
def add(n,k):
    print(n+k)
def sub(n,k):
    print(n-k)

def display(main,n,k):
    main(n,k)

display(add,x,y)
```
