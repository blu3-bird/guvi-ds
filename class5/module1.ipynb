```python
class wallet:
    hundred: 4
    fiveHundred: 2

    def __init__(self, hundred, fiveHundred):
        self.hundred = hundred
        self.fiveHundred = fiveHundred

    def total_cash(self):
        self.amount = (self.hundred * 100) + (self.fiveHundred * 500)


myWallet = wallet(1, 4)
myWallet.total_cash()
myWallet.__dict__
```

```python
marks = [20, 24, 30, 4]
max_marks = 30
percentage = []

for i in range(len(marks)):
    value = (marks[i] / max_marks) * 100
    percentage.append(value)

percentage
```

```python
percentage = [item / max_marks * 100 for item in marks]
percentage
```
