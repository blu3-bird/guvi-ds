### Section 1: If / Conditional Statements (Q1–Q5)


##### Q1. Write a program to check if a number is positive, negative, or zero.


```python
N = int(input())

if N < 0:
    print('-ve')
elif N == 0:
    print('zero')
else:
    print('+ve')
```

##### Q2. Check whether a given number is even or odd.


```python
N = int(input())

if N % 2 == 0:
    print('even')
else:
    print('odd')
```

##### Q4. Check if a year is a leap year.
(Hint: divisible by 4 but not 100 unless divisible by 400)


```python
N = int(input())

if N % 4 == 0:
    print('leap year')
else:
    print('normal year')
```

##### Q5. Write a program to assign grades based on marks:
90+ → A


75–89 → B


50–74 → C


<50 → Fail




```python
N = int(input())

grade = 'Fail'
if N >= 90:
    grade = 'A'
elif N <= 89 and N > 74:
    grade = 'B'
elif N <= 74 and N > 50:
    grade = 'C'


print(grade)
```

### Section 2: For Loop (Q6–Q10)


#### Q6. Print numbers from 1 to 20 using a for loop.


```python
N = int(input())
results = []

for i in range(1,(N+1),1):
    results.append(i)
print(*results)
```

```python
# using while loop
N = int(input())
results = []

i = 1
while i < (N+1):
    results.append(i)
    i +=1

results
```

#### Q7. Print all even numbers between 1 and 50.


```python
N = int(input())
results = []
for i in range(N):
    if i % 2 == 0:
        results.append(i)

print(results)
```

```python
# using while loop
N = int(input())
results = []
i = 0
while i < N:
    if i % 2 == 0:
        results.append(i)
    i +=1

print(*results)
```

#### Q8. Find the sum of first N natural numbers.

```python
N = int(input())

results = []
for i in range(N):
    results.append(i)

print(sum(results))
```

#### Q9. Print the multiplication table of a given number.


```python
N = int(input())

results = []

for i in range(1,11):
    product = N * i
    results.append(product)
print(results)
```

#### Q10. Count the number of vowels in a string.


```python
S = input()
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
count = 0
for char in S:
    if char in vowels:
        count +=1

print(count if count else 'no vowels')
```

#### Section 3: While Loop (Q11–Q14)


#### Q11. Print numbers from 10 to 1 using a while loop.


```python
N = int(input())
i = 1
results = []

while i < N+1:
    results.append(i)
    i +=1
results.sort(reverse=True)
print(results)
```

#### Q12. Find the factorial of a number using while loop.

```python
N = int(input())

product = 1
i = 1

while i < N+1:
    product *=i

    i +=1
print(product)
```

#### Q13. Keep asking the user for input until they enter “exit”.


```python
S = input()

while S:
    if S == 'exit':
        break
    else:
        S = input()
```

#### Q14. Reverse a number using while loop.


```python
N = input()

i = 0
results = []
while i < len(N):
    results.append(N[i])
    i +=1
results.reverse()
print(*results)

```
