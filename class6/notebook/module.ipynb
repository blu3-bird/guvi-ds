##### Map function

```python
scores = [33, 33, 12, 40]
max_marks = 40

percentage = list(map(lambda x: x * 100 / max_marks, scores))
percentage
```

##### filter function

```python
scores = [23, 40, 33, 12, 35]
max_marks = 40

passed_marks = list(filter(lambda x: x >= 33, scores))
passed_marks
```

```python
lst = list(map(int, input().split()))

num = list(filter(lambda x: x % 2 == 0, lst))
num_sq = list(map(lambda x: x**2, num))
num_sq
```

```python
def sum_arson(a, b):
    add = a + b
    sub = b - a

    return {"add": add, "sub": sub}


results = sum_arson(10, 23)
type(results)
```

```python
lst = [12, 14, 26, 34, 22, 45, 11]
num = 14

lst.sort()

low, high = 0, len(lst)
while low < high:
    mid = (low + high) // 2

    if mid > num:
        low = mid - 1
    elif mid < num:
        high = mid + 1
    else:
        print(mid)
```

```python
results = []
results_b = []
ans = []


def hcm(a, b):
    for i in range(1, a):
        if a % i == 0:
            results.append(i)
    for i in range(1, b):
        if b % i == 0:
            results_b.append(i)
        ans = results + results_b
        ans.sort()
    return ans


hcm(22, 40)
```
