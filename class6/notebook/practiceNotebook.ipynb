### Exercise 1.1: Precision Temperature Converter
**Objective:** Build a robust temperature conversion script.
**Requirements:**
- Prompt the user for a numeric temperature value and a unit (`C` or `F`)
- Validate input: reject values below absolute zero (`< -273.15°C` or `< -459.67°F`)
- Convert using standard formulas:  
  `F = (C × 9/5) + 32`  
  `C = (F − 32) × 5/9`
- Output the result formatted to exactly 2 decimal places

**Expected I/O:**
```
Enter temperature value: 36.6
Enter unit (C/F): C
Converted temperature: 97.88°F
```

````python
def temp_convertor():
    """Validate Input and then convert its units"""

    try:
        user_temp = float(input('Enter your temp: '))
    except ValueError:
        print('Invalid input.') 
        return
    user_unit = input('Enter temp Unit (C/F)').strip().lower()

    while user_unit not in ['c','f']:
        user_unit = input('Enter Temp unit: ').strip().lower()
    while True:
        if user_unit == 'c' and user_temp < -273.15:
            print('Invalid Input')
            user_temp = float(input('Enter Temp: '))
        elif user_unit == 'f' and user_temp < -459.67:
            print('Invalid Input')
            user_temp = float(input('Enter the Temp: '))
        else:
            break

    results = 0.0
    target_unit = ''

    if user_unit == 'c':
        results = (user_temp * 9/5) + 32
        target_unit = 'F'
    elif user_unit == 'f':
        results = (user_temp - 32) * 5/9
        target_unit = 'C'
    print(f'{results:.2f} {target_unit}')

if __name__ == '__main__':
    temp_convertor()
        
````

### Exercise 1.2: String Parser & Formatter
**Objective:** Manipulate and standardize user input strings.
**Requirements:**
- Accept a full name string (e.g., `"  dr. john A. smith  "`)
- Strip leading/trailing whitespace
- Split into first, middle (if present), and last components
- Output format: `"Smith, John A."` (Title-cased, last name first, comma-separated)
- Handle missing middle names gracefully without raising index errors

---

````python
def string_formatter(S:str)-> str:
    cleaned = S.strip()
    if not cleaned:
        return "Invalid Input"
    
    parts = cleaned.split()

    title_parts = [part.title() for part in parts]

    if len(title_parts) == 1:
        return title_parts[0]
    elif len(title_parts) == 2:
        return f'{title_parts[1]} {title_parts[0]}'
    else:
        first = title_parts[0]
        last = title_parts[-1]
        middle = title_parts[1:-1]
        middle_str = " ".join(middle)

        return f'{last}, {first} {middle_str}'

if __name__ == '__main__':
    s = input()
    print(string_formatter(s))
````

### Exercise 2.1: Academic Grade Classifier
**Objective:** Map numeric scores to letter grades with continuous user interaction.
**Requirements:**
- Accept numeric scores between 0–100
- Map to grades: A (90–100), B (80–89), C (70–79), D (60–69), F (<60)
- Return `"Invalid Score"` for out-of-range or non-numeric inputs
- Run inside a `while` loop until the user types `"quit"`

````python
def grade_classifer(score:int) -> str:
    """Academic grade classifer"""
    score = input('Enter your score: ')
    while score.strip().lower() != 'quit':
        max_marks = 100
        grade = ''
        if score.strip().lower() == 'quit':
            break
        elif int(score) not in range(0,(max_marks+1)):
             return 'Invalid Score'

        else:
            if int(score) in range(90,100):
                grade = 'A'
            elif int(score) in range(80,89):
                grade = 'B'
            elif int(score) in range(70,79):
                grade = 'C'
            elif int(score) in range(60,69):
                grade = 'D'
            elif int(score) < 60:
                grade = 'Fail'
        return grade

    if __name__ == '__main__':
        num = input()
        print(grade_classifer(num))
````

````python
def score_classifer(score:int) -> str: 
    """Grade Classifer"""
    grade = ''
    if score >=90:
        grade = 'A'
    elif score >=80:
        grade = 'B'
    elif score >=70:
        grade = 'C'
    elif score >=60:
        grade = 'D'
    else:
        grade = 'Fail'
    
    return grade


def main():
    while True:
        user_score = input('Enter your score: ')

        if user_score.strip().lower() == 'quit':
            print('Exiting the program')
            break

        try:
            score = float(user_score)
        except ValueError:
            print('Invalid Score')
            continue # instead of breaking program, taking the input again.

        if not 0<score<100:
            print('Invalid Score')
            continue
        
        else:
            print(score_classifer(score))

if __name__ == '__main__':
    main()
````


### Exercise 2.2: Optimized Prime Checker
**Objective:** Implement an efficient primality test.
**Requirements:**
- Write `is_prime(n)` returning `True`/`False`
- Handle edge cases: `n ≤ 1` (not prime), `2` (prime), even numbers (>2, not prime)
- Loop only up to `int(n**0.5) + 1` to check for divisors
- Include a driver loop that tests and prints all primes from 1 to 50

---


````python
def is_prime(num:int) -> bool:
    """Prime Checker"""

    if num <=1:
        return False
    elif num == 2:
        return True
    elif num % 2 == 0:
        return False
    else:
        limit = int(num**0.5) +1
        for i in range(3,limit,2):
            if num % i == 0:
                return False
        return True

if __name__ == '__main__':
    for i in range(0,51):
        if is_prime(i):
            print(i, end=' ')
    print()
````

### Exercise 3.1: Flexible Palindrome Validator
**Objective:** Create a configurable string validation function.
**Requirements:**
- Define `is_palindrome(text: str, ignore_case: bool = True, ignore_punctuation: bool = True) -> bool`
- Strip punctuation and normalize case based on boolean flags
- Return `True` if cleaned string reads identically forward and backward
- Include a comprehensive docstring and type hints


````python

def is_palindrome(text:str,ignore_case:bool =True, ignore_punctuation:bool = True) -> bool:
    """
        Palindrome Validator
        output: Return True if cleaned strings indentically forward and backward
    """

    cleaned = text.strip()

    if ignore_case:
        cleaned = cleaned.lower()
    
    if ignore_punctuation:
        cleaned = "".join(char for char in cleaned if char not in string.punctuation)
    
    cleaned = cleaned.replace(" ",'')
    
    return  cleaned == cleaned[::-1]


if __name__== '__main__':
    S = input()
    print(is_palindrome(S))
````

### Exercise 3.2: Dynamic Argument Aggregator
**Objective:** Master flexible function signatures.
**Requirements:**
- Define `aggregate_stats(*args: float, operation: str = "sum", **kwargs) -> float`
- Support operations: `"sum"`, `"mean"`, `"max"`, `"min"`
- Use `**kwargs` to handle optional flags like `ignore_zeros: bool = True`
- Raise `ValueError` for unsupported operations or empty positional arguments


````python
def aggregate_stats(*args:float, operation:str = 'sum', **kwargs) -> float:
    """
        Perform basic math operations

        args: list of float
        operation: list of some valid operations like sum,max,etc
        **kwargs: keyword args like ignore_zeros (flags)

        return: float
    """

    if not args:
        raise ValueError('Invalid Input,No numbers are given')
    
    ignore_zero = kwargs.get('ignore_zeros' , True)

    num = list(args)
    num_lst = []
    if ignore_zero:
        num_lst = [i for i in num if i !=0]
    
    if not num_lst:
        raise ValueError('Invalid Input, Only Zeros were given')
    
    valid_operations = ['sum','min','max','mean']

    if operation not in valid_operations:
        raise ValueError('Invalid Operation given')
    
    if operation == 'sum':
        results = sum(num_lst)
    elif operation == 'min':
        results = min(num_lst)
    elif operation == 'max':
        results = max(num_lst)
    elif operation == 'mean':
        results = sum(num_lst)/len(num_lst)
    
    return results

if __name__ == '__main__':
    print(aggregate_stats(10,operation = 'mean'))
````

#### Exercise 4.1: Word Frequency Analyzer
**Objective:** Process text and count word occurrences.
**Requirements:**
- Accept a paragraph string
- Convert to lowercase, remove punctuation, split into words
- Count frequencies using a dictionary
- Output words sorted by frequency (descending), then alphabetically for ties
---

````python
import string
def word_frequecy(txt:str) -> dict:
    """
    Obj: Process text and count word occurrences
    args: txt: str
    output: word : occurrences
    """
    
    txt_l = txt.lower()
    
    # words = txt_l.split()

    cleaned_txt = "".join(i for i in txt_l if i not in string.punctuation)

    
    words = cleaned_txt.split()
    count = {}
    for word in words:
        count[word] = count.get(word,0) +1

    sort_dict = sorted(count.items(), key= lambda item: (-item[1], item[0]))

    return sort_dict


if __name__ == '__main__':
    s = input('enter your text: ')
    results = word_frequecy(s)

for word, count in results:
    print(f'{word}: {count}')

````

### Exercise 4.2: Set-Based Data Filter
**Objective:** Leverage set theory for efficient list operations.
**Requirements:**
Given:
```python
list_a = [1, 2, 2, 3, 4, 5]
list_b = [4, 5, 5, 6, 7, 8]
```
- Convert to sets to remove duplicates
- Compute intersection (`&`) and symmetric difference (`^`)
- Return results as sorted lists without duplicates
---

````python
def set_theory(a:list,b:list)->list:
    """
    Obj: Leverage Set Theory for efficient list operations
    arg: a-> list, b-> list
    remove duplicates, and return sorted list.
    return: list
    """


    a_set = set(a)
    b_set = set(b)

    symmetric_set = a_set ^ b_set
    intersection_set = a_set & b_set
    return {
        'Intersection': sorted(list(intersection_set)),
        'symmetric':sorted(list(symmetric_set))
    }
lst = [2,4,6,8,10]
lst_1 = [2,4,6,8]
print(set_theory(lst,lst_1))

    
````

### Exercise 5.1: Structured Log Parser
**Objective:** Safely read, parse, and summarize external data.
**Requirements:**
- Read `access.log` (format: `TIMESTAMP IP STATUS_CODE MESSAGE`)
- Count HTTP status code occurrences
- Write summary to `summary.txt` as `STATUS_CODE: COUNT`
- Use `with open(...)` for deterministic resource management
- Handle `FileNotFoundError` and malformed lines gracefully

---



````python
def log_parser(input_file:str ='access.log',output_file:str='summary.txt')->None:
    """
        Read access.log file and read the occerrences in the of HTTP status code
        return: summary.txt file contains the occrrences of HTTP 
    """
    status_code = {}
    try:
        with open(input_file,'r',encoding='utf-8') as f:
            for line_num, line in enumerate(f,start=1):

                line = line.strip()
                if not line:
                    print(f'Invalid Line {line_num}:{line}')
                    continue

                parts = line.split()

                if len(parts) <3:
                    print(f'Invalid Line {line_num}:{line}')
                    continue
                
                code = parts[2]

                if not code.isdigit() or len(code) != 3:
                    print('Invalid Code')
                    continue

                status_code[code] = status_code.get(code,0)+1
                   
    except FileNotFoundError:
        print('Access.log file not found','error')
        return
    except Exception as e:
        print(f'Error: {e}')
        return
    

    # open a summary.txt file and adding text there.
    try:
        with open(output_file,'w',encoding='utf-8') as out:
            for code in sorted(status_code.keys()):
                out.write(f'{code}: {status_code[code]}\n')
    except Exception as e:
        print(f'Error:{e}')
        return


if __name__ == '__main__':
    log_parser()
````

````python
def file_read(input_file:str='access.log',output_file:str='summary.txt')->None:
    """Read a file , extract text from it and writes in a different file"""
    status_count = {}
    try:
        with open(input_file,'r',encoding='utf-8') as f:
            for line_num, line in enumerate(f,start=1):

                line = line.strip()

                if not line:
                    print(f'Empty line {line_num}:{line}')
                    continue

                parts = line.split()

                if len(parts)<3:
                    print(f'Length of line is less than 3(invalid line) {line_num}:{line}')
                    continue
                
                code = parts[2]

                if not code.isdigit() or len(code)<3:
                    print(f'Invalid line(code is invalid) {line_num}:{line}')
                    continue


                # storing status code in dict.
                status_count[code] = status_count.get(code,0) +1

    except FileNotFoundError:
        print(f'{input_file} not found, check folder path')
        return
    except Exception as e:
        print(f'Error at {e}')

    try:
        with open(output_file,'w',encoding='utf-8') as w:
            for code in sorted(status_count.keys()):
                w.write(f'{code}: {status_count[code]}\n')
    except Exception as e:
        print(f'Error {e}')
        return
    
if __name__ == '__main__':
    file_read()
                
````
