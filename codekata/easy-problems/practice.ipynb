Sample Input:

### 10. Check if Number is Composite

Problem Statement:
Given a number N, print 'yes' if it is composite else print 'no'.


Sample Input:
123


Sample Output:
yes

```python
N = 1 # input field
N = str(N)  # converting int to str to call each elemets in it.

is_composite = False # setting default value

if int(N) <= 1: # 0 and 1 are not composite number
    print('no')
else:
    for char in range(2,int(N)): # calling all the char in the range of 2 to n
        if int(N) % char == 0: # suppose char = 20, 
            is_composite = True
            break
        else:
            is_composite = False
    
    if is_composite:
        print('yes')
    else:
        print('no')
```

#### 2. Prime Number Check


Problem Statement:
Given a number N, check whether it is prime or not. Print 'yes' if it is prime else print 'no'.


Input Description:
The input consists of a single integer N.


Output Description:
The output is 'yes' if N is prime, otherwise 'no'.


Sample Input:
123


Sample Output:
no

```python
N = 17 # taking int input

if N <= 1: # if input is one or less than one, 
    print('No')
elif N == 2: # is N is 2, then call it prime and break the loop
    print('Yes')
elif N % 2 == 0: # for all the even cases
    print('No')
else: # for all the cases,
    is_prime = True # initial setting is_prime = True
    for char in range(3,int(N ** 0.5)+1,2): #looping in the range from 3 to sqrt of N, why 3?, cause we already covered 1,2 in above cases, syntax: start, end, step. (why step is 2, to avoid even cases,) why plus 1 in end point, to cover all the possible odd cases.
        if N % char == 0: # is N is divided by those char, then is_prime is false, 
            is_prime = False
            break # breaking the loop
    print('Yes' if is_prime else 'No') # print statement.
```

```python
# same solution for above question but with func

N = 10

def chk_prime(num):
    if num <=1:
        print('no')
    elif num == 2:
        print('yes')
    elif num % 2 == 0:
        print('yes')
    else:
        is_prime = True
        for digit in range(3, int(num**0.5)+1,2):
            if num % digit == 0:
                is_prime = False
                break
        print('yes' if is_prime else 'no')

chk_prime(N)
```

```python
N = 10 # taking input

def chk_prime(num):
    if num <=1:
        return False # return False, if check is true
    else:
        for digit in range(2,int(num**0.5)+1): # covering even cases too, 
            if num % digit == 0:
                return False
        return True # if loops return True

chk_prime(N)
```

```python
# same solution but using any() func

N = 10

if N <=1:
    print('no')
elif N == 2:
    print('yes')
elif N % 2 == 0:
    print('no')
else:   
    is_prime = True
    if any(N % digit == 0 for digit in range(3, int(N**0.5)+1,2)):
        is_prime = False
        
    print('yes' if is_prime else 'no')
```

####  Next Greater Power of 2


Question:
Given a number N, find its next immediate greater power of 2(i.e 2^1, 2^2, 2^3...).


Input Description:
The input consists of a number N where N <= 1000.


Sample Input:
4


Sample Output:
8



```python
N = 31

exp = 0 # exponetation 
while (2 ** exp) <= N:  # exponetation starting from zero, 2 power exp exceeds the input N, 
    exp = exp + 1 # when loops break, we add new exp and print it out.

print(2 ** exp)
```

#### 8. Round to Nearest Greater Integer


Question:
Given a floating point number with 1 decimal place round it off to nearest greater integer and print it.


Sample Input:
2.6


Sample Output:
3



```python
N = 6.9 # taking input, as float

NInt = int(N) # converting float to int, 6.9 to 6 if it is 8.0 then it will become 8 only

if N > NInt: # checking if float is greater than int, then add one otherwise print int
    print(NInt+1)
else:
    print(NInt)
```

#### 11. Print Factors of a Number


Problem Statement:
Given a number N, print its factors.


Input Description:
Input Size : n<=1000


Sample Input:
6


Sample Output:
1 2 3 6

```python
N = 34 # taking int input

num = [] # list to store factors
for i in range(1,N+1): # to find factors, we need divide every number for 1 to N, first running loops,
    if N % i == 0: # numbers leave 0 remainder on dividing, adding those in list
        num.append(i)
print(*num) # print only elements of list, * print only elements, and add space in between them.
```

```python
# alternative solution

N = 65
for i in range(1,N+1):
    if N % i == 0:
        print(i, end=" ") # simply printing i, using end method, which adds space after every elements
        # print(i)
```

#### 13. Perfect Square Product


Problem Statement:
Given 2 numbers N,M. Print 'yes' if their product is a perfect square else print 'no'.


Sample Input:
5 5


Sample Output:
yes



```python
# perfect sq means, 16 which 4 * 4, so here 16 is perfect sq, however 10 which 2 * 5 , so ten cannot be a perfect sq, as for perfect sq, sq root of number must be a whole number, like 16 

N = '9 9'

num = N.split()

N = int(num[0])
M = int(num[1])

product = N * M # product of provided numbers

root = product ** 0.5 # taking root of the product, if the root is whole number, then product was perfect sq.

if root % 1 == 0: # if a number is completely divided by 1, then it is a whole number, 
    print('yes')
else:
    print('no')
```
