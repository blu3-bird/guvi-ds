#### Question 23:- Product of Digits


Problem Statement:
Given a number N, print the product of the digits.


Input Description:
Input Size : N <= 100000000000


Sample Input:
2143

Sample Output:
24


```python
N = '5'

results = 1
for digit in N:
    results *= int(digit)
print(results)
```

Question:
Given a string S, print it after changing the middle element to * (if the length of the string is even, change the 2 middle elements to *).


Sample Input:
hello


Sample Output:
he*lo


```python
S = 'fghjkl;;'
mid = len(S) // 2 # // will return rounded off answer,(in odd case number)

if len(S) % 2 == 1: # idd casem
    ans = S[:mid] + '*' + S[mid+1:] # S[starting point end point]
elif len(S) % 2 == 0:
    ans = S[:mid-1] + '**' + S[mid+1:] # [1 2 * * 5 6] mid = 3

print(ans)
```

### 10. Check if Number is Composite

Problem Statement:
Given a number N, print 'yes' if it is composite else print 'no'.


Sample Input:
123


Sample Output:
yes

```python
N = 1 # input field
N = str(N)  # converting int to str to call each elemets in it.

is_composite = False # setting default value

if int(N) <= 1: # 0 and 1 are not composite number
    print('no')
else:
    for char in range(2,int(N)): # calling all the char in the range of 2 to n
        if int(N) % char == 0: # suppose char = 20, 
            is_composite = True
            break
        else:
            is_composite = False
    
    if is_composite:
        print('yes')
    else:
        print('no')
```

#### 2. Prime Number Check


Problem Statement:
Given a number N, check whether it is prime or not. Print 'yes' if it is prime else print 'no'.


Input Description:
The input consists of a single integer N.


Output Description:
The output is 'yes' if N is prime, otherwise 'no'.


Sample Input:
123


Sample Output:
no

```python
N = 17 # taking int input

if N <= 1: # if input is one or less than one, 
    print('No')
elif N == 2: # is N is 2, then call it prime and break the loop
    print('Yes')
elif N % 2 == 0: # for all the even cases
    print('No')
else: # for all the cases,
    is_prime = True # initial setting is_prime = True
    for char in range(3,int(N ** 0.5)+1,2): #looping in the range from 3 to sqrt of N, why 3?, cause we already covered 1,2 in above cases, syntax: start, end, step. (why step is 2, to avoid even cases,) why plus 1 in end point, to cover all the possible odd cases.
        if N % char == 0: # is N is divided by those char, then is_prime is false, 
            is_prime = False
            break # breaking the loop
    print('Yes' if is_prime else 'No') # print statement.
```

```python
# same solution for above question but with func

N = 10

def chk_prime(num):
    if num <=1:
        print('no')
    elif num == 2:
        print('yes')
    elif num % 2 == 0:
        print('yes')
    else:
        is_prime = True
        for digit in range(3, int(num**0.5)+1,2):
            if num % digit == 0:
                is_prime = False
                break
        print('yes' if is_prime else 'no')

chk_prime(N)
```
