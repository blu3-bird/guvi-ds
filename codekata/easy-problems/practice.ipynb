#### Question 23:- Product of Digits


Problem Statement:
Given a number N, print the product of the digits.


Input Description:
Input Size : N <= 100000000000


Sample Input:
2143

Sample Output:
24


```python
N = '5'

results = 1
for digit in N:
    results *= int(digit)
print(results)
```

Question:
Given a string S, print it after changing the middle element to * (if the length of the string is even, change the 2 middle elements to *).


Sample Input:
hello


Sample Output:
he*lo


```python
S = 'fghjkl;;'
mid = len(S) // 2 # // will return rounded off answer,(in odd case number)

if len(S) % 2 == 1: # idd casem
    ans = S[:mid] + '*' + S[mid+1:] # S[starting point end point]
elif len(S) % 2 == 0:
    ans = S[:mid-1] + '**' + S[mid+1:] # [1 2 * * 5 6] mid = 3

print(ans)
```

### 10. Check if Number is Composite

Problem Statement:
Given a number N, print 'yes' if it is composite else print 'no'.


Sample Input:
123


Sample Output:
yes

```python
N = 1 # input field
N = str(N)  # converting int to str to call each elemets in it.

is_composite = False # setting default value

if int(N) <= 1: # 0 and 1 are not composite number
    print('no')
else:
    for char in range(2,int(N)): # calling all the char in the range of 2 to n
        if int(N) % char == 0: # suppose char = 20, 
            is_composite = True
            break
        else:
            is_composite = False
    
    if is_composite:
        print('yes')
    else:
        print('no')
```
