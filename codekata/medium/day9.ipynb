##### 45. Sorted Elements Less Than K


Problem Statement:
Given 2 numbers N,K followed by N elements print all the elements lesser than K in sorted order.If the elements could not be found print -1


Input Description:
Input Size : N <= 100000


Sample Input:
5 3 

1 2 1 4 1


Sample Output:
1 1 1 2



```python
# question: given two numbers, and an array. we need to print the elements which are less than second input. if there is no print -1

N, K = map(int, input().split())  # taking input from user,
arr = list(map(int, input().split()))  # taking input as list.

data = []  # empty list to store elements.

for i in arr:
    if i < K:
        data.append(i)

if data:
    data.sort()  # sorting list.
    print(*data)
else:
    print(-1)
```

##### 46. Descending Elements Below N


Problem Statement:
Given a number N and an array of N elements, print all elements lesser than N in descending order. If no element found print -1.


Input Description:
The input consists of a number N, and an array of N elements. N is between 1 and 10000 (inclusive).


Output Description:
Print all elements from the array that are lesser than N, in descending order. If no such elements are found, print -1.


Sample Input:
5

2 14 15 14 3


Sample Output:
3 2

```python
N = int(input())  # number of elements in arr

arr = list(map(int, input().split()))

i = 0  # defining i = 0
data = []

while i < N:  # i < number of elements in arr
    if arr[i] < N:
        data.append(arr[i])
    i = i + 1  # adding 1 in i to exit the while loop

if data:
    data.sort(reverse=True)  # sorting list in descending order
    print(*data)
else:
    print(-1)
```

##### 47. Numbers Repeated K Times


Problem Statement:
Given 2 numbers N,K and an array of N elements, print the number(s) that has been repeated K times.Print them in ascending order if there are more than one number to be printed.If no element satisfies the pattern then print -1


Input Description:
The input consists of two integers N and K, followed by an array of N elements. N and K are up to 100000.


Output Description:
Print the numbers that have been repeated K times in ascending order. If no such element exists, print -1.


Sample Input:
5 2

1 2 4 1 2


Sample Output:
1 2



```python
N, K = map(int, input().split())
arr = list(map(int, input().split()))

count = {}

for num in arr:
    if num in count:
        count[num] = count[num] + 1
    else:
        count[num] = 1

results = []
for num, freq in count.items():
    if freq == K:
        results.append(num)

if results:
    results.sort(reverse=True)
    print(*results)
else:
    print(-1)
```

```python
# set method
N, K = map(int, input().split())
arr = list(map(int, input().split()))

unique = set(arr)
results = []

for num in unique:
    if arr.count(num) == K:
        results.append(num)

if results:
    results.sort(reverse=True)
    print(*results)
else:
    print(-1)
```

##### 49. Longest Repeating Sequence


Problem Statement:
Given a number N and an array of N elements, find the length of the longest repeating sequence of the elements.If no such sequence is found print -1


Input Description:
The input consists of an integer N (where N <= 100000) and an array of N elements.


Sample Input:
8

1 2 2 2 3 4 5 6


Sample Output:
3

```python
N = int(input())
arr = list(map(int, input().split()))

count = 1
max_length = 1

for i in range(len(arr)):
    if arr[i] == arr[i - 1]:
        count = count + 1
        max_length = max(count, max_length)
    else:
        count = 1

if max_length > 1:
    print(max_length)
else:
    print(-1)
```

##### Question3: Given N numbers, print sum of elements at even indices (0, 2, 4...).

Input:
5

10 20 30 40 50

Output:
90

(10 + 30 + 50 = 90)

```python
N = int(input())
arr = list(map(int, input().split()))
sum = 0


for i in range(N + 1):
    if i % 2 == 0:
        sum = sum + arr[i]
print(sum)
```

```python
# alternative way
N = int(input())
arr = list(map(int, input().split()))

sum = 0
for i in range(0, N, 2):
    sum = sum + arr[i]
print(sum)
```

#### SECTION 2: Frequency & Counting

##### Question4: Given N numbers, print the most frequent element.
If tie, print smallest number.

> Input:
7

1 2 2 3 3 3 1

> Output:
3

(3 appears 3 times)


```python
N = int(input())
arr = list(map(int, input().split()))

count = {}

for num in arr:
    if num in count:
        count[num] = count[num] + 1
    else:
        count[num] = 1

# finding the max frequency
max_freq = max(count.values())

results = []
for num, freq in count.items():
    if freq == max_freq:
        results.append(num)

if len(results) < 2:
    print(min(results))
else:
    print(min(arr))
```

##### Question6: Given numbers from 1 to N with one missing, find the missing number.

Input:
5

1 2 4 5

Output:
3

```python
N = int(input())
arr = list(map(int, input().split()))

actual_sum = 0
natural_sum = 0
for num in arr:
    actual_sum = actual_sum + num

n_arr = []
for i in range(1, (N + 1)):
    n_arr.append(i)

for num in n_arr:
    natural_sum = natural_sum + num


print(natural_sum - actual_sum)
```

```python
# alternative way:
N = int(input())
arr = list(map(int, input().split()))

natural_arr = []
for i in range(1, (N + 1)):
    natural_arr.append(i)


print((sum(natural_arr)) - (sum(arr)))
```

```python
# alternative way to solve this
N = int(input())
arr = set(map(int, input().split()))

n_arr = set()

for i in range(1, (N + 1)):
    n_arr.add(i)

set(n_arr)

missing = (n_arr - arr).pop()
print(missing)
```

#### SECTION 3: Two Pointers & Pairs

##### Question7: Given N numbers and K, check if any two numbers differ by K.

Input:
5 3

1 5 3 4 2

Output:
yes

(5-2=3 or 4-1=3)

```python
N, K = map(int, input().split())
arr = list(map(int, input().split()))

found = False
for i in range(N):
    for j in range(i + 1, N):
        if arr[i] - arr[(j)] == K:
            found = True
            break
    if found:
        break


if found:
    print("yes")
else:
    print(-1)
```

```python
# alternative method
N, K = map(int, input().split())
arr = list(map(int, input().split()))

found = False
for i in arr:
    if (K + i) in arr:
        found = True

if found:
    print("yes")
else:
    print("no")
```
