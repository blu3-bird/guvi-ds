##### 34. Kth Smallest Element

Problem Statement:
Given 2 numbers N and K followed by N elements, find the Kth smallest element. If the element cannot be found then print -1


Input Description:
The input consists of two numbers N and K, followed by N elements. N <= 100000.


Output Description:
The output is the Kth smallest element. If the element cannot be found, print -1.


Sample Input:

5 2

1 1 2 4 5


Sample Output:
2

```python
N = 5
K = 4
data = [3, 2, 5, 5, 9]

num = list(set(data))
num.sort()

if len(num) < K:
    print("-1")
else:
    print(num[K - 1])
```

##### 37. First Occurrence of Character


Problem Statement:
Given a string 'S' and a character 'K', find at what position the character 'K' occurs for the first time in 'S'.(Assume the index of string starts at 1).If the character is not found in 'S' then print -1


Input Description:
Input Size : |s| <= 100000


Sample Input:
codekata a


Sample Output:
6



```python
S = "arsonarson"
K = "a"

found = False
count = 0

for char in S:
    count = count + 1
    if char == K:
        found = True
        result = count
        break

if found:
    print(result)
else:
    print(-1)
```

```python
S = "gfjkdgfadfgvafdas"
K = "g"

found = False
for idx, char in enumerate(S):
    if char == K:
        found = True
        print(idx + 1)
        break
else:
    print(-1)
```

##### 39. Count String Occurrences


Problem Statement:
Given a sentence and string S, find how many times S occurs in the given sentence.If S is not found in the sentence print -1


Input Description:
Input Size : |sentence| <= 1000000(complexity O(n)).


Output Description:
The output is the number of times S occurs in the given sentence, or -1 if S is not found.


Sample Input:
I enjoy doing codekata
codekata


Sample Output:
1

```python
S = "Arson Arson Arson Everything bruh"  # taking input statement as str
K = "Arson"

S_list = S.split()  # splitting statement in elements.

found = False
count = 0
for char in S_list:
    if char == K:
        found = True
        count = count + 1


if found:
    print(count)
else:
    print(-1)
```

##### 42. Pair Sum Check


Problem Statement:
Given 2 numbers N,X and an array of N elements, check if there exists any 2 numbers in the array with sum equal to X.If found print 'yes' otherwise print 'no'


Input Description:
The input consists of two numbers N and X, and an array of N elements. N and X are up to 100000.


Output Description:
Print 'yes' if two numbers with sum equal to X are found in the array, otherwise print 'no'.


Sample Input:
4 4

2 2 0 0


Sample Output:
yes

```python
# question: sum of any two elements in arr is equal to X
N = 6  # number of elements in array
X = 1

arr = [1, 4, 3, 0, 4, 1]  # array

found = False  # bool to track, whether the sum is equal to X

for i in range(N):  # looping through the index of our array.
    for j in range(
        (i + 1), N
    ):  # inner looping, but with starting point (i+1), which means that if i = 0, j is 1.
        if arr[i] + arr[j] == X:
            found = True
            break  # once found = True, break the inner
    if found:  # if in inner loop found is True, breaking the outer loop too.
        break

if found:
    print("yes")
else:
    print("no")
```

```python
# alternative method set()

N = 5
X = 4
arr = [2, 4, 3, 16, 3]

seen = set()  # set to store arr elements
found = False

for i in arr:  # looping through arr
    toAdd = X - i  # toAdd var defines how much we need to add, to get output X.
    if toAdd in seen:  # checking toAdd is in set.
        found = True
        break
    seen.add(i)  # if we found toAdd in seen, then break otherwise add i in seen.

if found:
    print("yes")
else:
    print("no")
```

##### 43. Minimum Odd Quotient Factor


Problem Statement:
Given number N, find the minimum factor which yeilds odd number as the quotient.


Input Description:
The input consists of a single integer N, where N <= 100000.


Output Description:
The output is the minimum factor of N that yields an odd number as the quotient.


Sample Input:
9


Sample Output:
1



```python
N = 220  # taking input
factor = []  # list to store factors

for i in range(1, (N + 1)):  # looping from 1 to N+1, as range fucn exclude end point
    if N % i == 0:  # if number completely divides N then it is factor of N
        factor.append(i)

for num in factor:  # looping through factor
    if (
        (N // num) % 2 == 1
    ):  # N // num to get quotient, if quotient is divided by 2 and returns 1, then print that factor.
        print(num)
        break
```

```python
# altenative method
N = 220

for i in range(1, (N + 1)):
    if N % i == 0:
        quotient = N // i  # finding quotient, dividing N by its factor
        if quotient % 2 == 1:  # checking if quotient is odd or even
            print(i)  # print factor
            break  # as we find our fist match break the loop
```

##### 44. Common Elements in Sorted Arrays


Problem Statement:
Given a number N and 2 arrays A and B of sorted order of size N, print the common elements.If it is not found print -1.


Input Description:
Input Size : 1 <= N <= 100000


Sample Input:
5

1 1 1 1 1

1 2 3 4 5


Sample Output:
1



```python
N = 6
A = [1, 2, 3, 33, 1, 9]
B = [3, 3, 33, 9, 2, 2]  # taking input as list, instead we can directly take it as set.

A = set(A)
B = set(B)

common = []  # list to store common elements

for i in A:
    if i in B:
        common.append(i)

common.sort()  # sorting list, as order doesnt maintain any order
if common:
    print(common)
else:
    print(-1)
```

```python
# alternative method
N = 3
A = {2, 6, 3}
B = {1, 6, 5}

common = sorted(
    A & B
)  # sorted() to sort our set, & will pick common elements and add in our var.S

if common:
    print(*common)
else:
    print(-1)
```
