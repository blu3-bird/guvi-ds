##### 34. Kth Smallest Element

Problem Statement:
Given 2 numbers N and K followed by N elements, find the Kth smallest element. If the element cannot be found then print -1


Input Description:
The input consists of two numbers N and K, followed by N elements. N <= 100000.


Output Description:
The output is the Kth smallest element. If the element cannot be found, print -1.


Sample Input:

5 2

1 1 2 4 5


Sample Output:
2

```python
N = 5
K = 4
data = [3,2,5,5,9]

num = list(set(data))
num.sort()

if len(num) < K:
    print('-1')
else:
    print(num[K-1])
```

```python
N = 6 # number of elements in list
K = 6 # input of kth element

n_list = [1,3,4,2,5,5] # input list

# removing duplicates.
n_list = list(set(n_list)) # convertin to set, then to list again.

# sorting the list to find K smallest element in list.
n_list.sort()

if len(n_list) < K: # see per th question, we need to find the Kth smallest element in list. if the len of list is smaller than K, then there is no Kth element in list.
    print(-1)
else:
    print(n_list[K-1]) # print K-1 element, as in list, index starts from 0 not 1.(6th element means 5 element in list. so K-1)

```

##### 37. First Occurrence of Character


Problem Statement:
Given a string 'S' and a character 'K', find at what position the character 'K' occurs for the first time in 'S'.(Assume the index of string starts at 1).If the character is not found in 'S' then print -1


Input Description:
Input Size : |s| <= 100000


Sample Input:
codekata a


Sample Output:
6



```python
S = 'arsonarson'
K = 'a'

found = False
count = 0

for char in S:
    count = count + 1
    if char == K:
        found = True
        result = count
        break
    
if found:
    print(result)
else:
    print(-1)
```

```python
S = 'gfjkdgfadfgvafdas'
K = 'g'

found = False
for idx,char in enumerate(S):
    if char == K:
        found = True
        print(idx+1)
        break
else:
    print(-1)
```

```python
S = 'arson d ace' # string input
K = 'd' # character to search

for idx,char in enumerate(S): # looping through our string. nd taking its index, and char
    if char == K: # if any char matches with K, print its index.(index + 1 as per question.)
        print(idx+1)
        break   # breaking the looop

else: # outside of if statement, cause if we add this inside our if statement, it will print -1 for every no matching case.
    print(-1)
```

```python
# alernative solution for this
S = 'gfdsfgdgew'
K = 'x'

found = False # bool, initial set to false.

for idx, char in enumerate(S):
    if char == K:
        found = True
        index = idx
        break

if found:
    print(index+1)
else:
    print(-1)
```

##### 39. Count String Occurrences


Problem Statement:
Given a sentence and string S, find how many times S occurs in the given sentence.If S is not found in the sentence print -1


Input Description:
Input Size : |sentence| <= 1000000(complexity O(n)).


Output Description:
The output is the number of times S occurs in the given sentence, or -1 if S is not found.


Sample Input:
I enjoy doing codekata
codekata


Sample Output:
1

```python
S = 'Arson Arson Arson Everything bruh' # taking input statement as str
K = 'Arson' 

S_list = S.split() # splitting statement in elements. 

found = False
count = 0
for char in S_list:
    if char == K:
        found = True
        count = count + 1
    

if found:
    print(count)
else:
    print(-1)
```
