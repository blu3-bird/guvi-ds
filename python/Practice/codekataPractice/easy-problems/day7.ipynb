##### 32. Sum of First K Natural Numbers


Problem Statement:
Write a program to print the sum of the first K natural numbers.


Input Description:
Input Size : n <= 100000


Sample Input:
3


Sample Output:
6



```python
# using while loop

K = 12  # taking input as int

sum = 0  # defining a var with initial value 0.
while K > 0:  # loopping through all the from 1 to K.
    sum = sum + K  # Adding K in every loop in var
    K = (
        K - 1
    )  # as in loop, we going up by 1 digit in each loop, so we gonna minus -1 each time, our loop runs.

print(sum)  # printing sum, outside of our looop
```

```python
# using for loops
K = 4

sum = 0

for i in range(1, K + 1):  # looping through 1 to K,
    sum = sum + i  # adding each digit in sum. var

print(sum)
```

##### 33. Second Smallest Element

Problem Statement:
Given a number N followed by N elements, find the second smallest element.If it cannot be found then print -1


Input Description:
Input Size : N <= 100000 (ie do it in O(log n) time complexity)


Sample Input:
5
1 2 3 4 5


Sample Output:
2



```python
N = 5  # number of elements present in list.
num = [3, 3, 3, 3, 3]  # list

# removing duplicates by converting list to set, then set to list, again to sort it.
num_list = list(set(num))

num_list = sorted(num_list)  # sorting list.

if (
    len(num_list) < 2
):  # if len of list is less than 2, means there were four duplicates in list, so no second small element
    print("-1")
else:
    print(num_list[1])
```

##### 35. String Length Without Functions


Problem Statement:
Given a string S, find its length(including the spaces)without using any pre-defined functions.


Sample Input:
codekata


Sample Output:
8

```python
S = "arson D Ace"  # input().split() means it will split the input in words. rn program will consider the str as a collective word.

count = 0
for char in S:
    count = count + 1
print(count)
```

```python
# alernative
S = "jd   fhd"


def count(s):
    length = 0
    for char in s:
        length = length + 1

    return length


count(S)
```

##### 41. Common Characters in Strings


Problem Statement:
Given 2 strings,check whether they have any common characters.If found print 'yes' else print 'no'.


Input Description:
Input Size : |s| <= 100000(O(n))


Sample Input:
guvi guvigeeks


Sample Output:
yes



```python
# set method

S = "arson is crazy"
K = "d"  # if add space here, it will print yes, as set will create an empty element.

if set(S) & set(
    K
):  # Converting str to set, as it will duplicates, and keep only unique elements.
    print("yes")
else:
    print("no")
```

```python
# loop method

S = "arson everything"
K = "lfjakd"
found = False
for char in S:
    if char in K:
        found = True
        break

if found:
    print("yes")
else:
    print("no")
```
