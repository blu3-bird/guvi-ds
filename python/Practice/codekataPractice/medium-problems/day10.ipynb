##### Question 7: Given N numbers and K, check if any two numbers differ by K.

Input:
5 3

1 5 3 4 2

Output:
yes
(5-2=3 or 4-1=3)

```python
N, K = map(int, input().split())
arr = list(map(int, input().split()))

found = False
for i in range(N):
    for j in range(i + 1, N):
        if arr[i] - arr[j] == K:
            found = True
            break
    if found:
        break

if found:
    print("yes")
else:
    print("no")
```

##### Question 5: Given N numbers, print all elements that appear exactly once.
Print in ascending order. If none, print -1.

Input:
6

1 2 2 3 4 4

Output:
1 3

```python
N = int(input())
arr = list(map(int, input().split()))

count = {}

for i in arr:
    if i in count:
        count[i] = count[i] + 1
    else:
        count[i] = 1

results = []
for num, freq in count.items():
    if freq == 1:
        results.append(num)

if results:
    print(*results)
else:
    print(-1)
```

##### Question 10: Given two strings, print common vowels (a,e,i,o,u) in alphabetical order.
If none, print -1.

Input:

hello world

Output:
o

```python
S, K = input().split()

results = set()
vowels = ["a", "e", "i", "o", "u"]

for char in S:
    if char in K:
        if char in vowels:
            results.add(char)

if results:
    print(*results)
else:
    print(-1)
```

#### Question 13: Given N numbers, find length of longest consecutive increasing subsequence.

Input:
8

1 2 3 2 3 4 5 1

Output:
4

(2 3 4 5)

```python
N = int(input())
arr = list(map(int, input().split()))


current_length = 1
max_length = 1

for i in range(N):
    if arr[i] > arr[i - 1]:  # checking if current element is greater than previous
        current_length += 1
    else:
        current_length = 1

    max_length = max(
        current_length, max_length
    )  # In each loop, if current_length is greater than max, it will be saved in max_length.

if max_length:
    print(max_length)
else:
    print(-1)
```

#### Question 14: Given N numbers, find longest alternating positive/negative sequence.

Input:
6

1 -2 3 -4 -5 6

Output:
4

(1 -2 3 -4)

```python
N = int(input())
arr = list(map(int, input().split()))

current_length = 0
max_length = 0

for i in range(1, N):
    if arr[i] * arr[(i - 1)] < 0:
        current_length += 1
    else:
        current_length = 1

    max_length = max(current_length, max_length)

if max_length:
    print(max_length)
else:
    print(-1)
```

#### Question 15: Given N numbers, count how many segments of consecutive zeros exist.

Input:
10

1 0 0 2 3 0 0 0 4 0

Output:
3
(two 0s, three 0s, one 0)

```python
N = int(input())
arr = list(map(int, input().split()))

current_length = 1
max_length = 1

for i in range(1, N):
    if arr[i] == 0 and arr[(i - 1)] == 0:
        current_length += 1
    else:
        current_length = 1

    max_length = max(current_length, max_length)

if max_length:
    print(max_length)
else:
    print(-1)
```

```python
# alternative method
N = int(input())
arr = list(map(int, input().split()))

current_length = 0
max_length = 0


def get_sign(x):
    if x == 0:
        return 1
    else:
        return 0


for i in range(1, N):
    if get_sign(arr[i]) and get_sign(arr[(i - 1)]) == 1:
        current_length += 1
    else:
        current_length = 1

    max_length = max(current_length, max_length)

if max_length > 1:
    print(max_length)
else:
    print(-1)
```

#### Question 18: Given N numbers and K, check if any consecutive subarray sums to K.

Input:
6 9

1 2 3 4 5 6

Output:
yes
(2+3+4=9)

```python
N, K = map(int, input().split())
arr = list(map(int, input().split()))

found = False
left = 0  # left window index
current_sum = 0

for (
    right
) in arr:  # adding all the elements in current sum, starting right index(0,1,2,..)
    current_sum += right

    while current_sum > K and left <= right:  # run while if current_sum > K,
        current_sum -= arr[left]  # current_sum - arr[0]
        left += 1  # inc. left index by 1

    if current_sum == K:
        found = True
        break

print("yes" if found else "no")
```

#### Question 19: A peak is an element greater than its neighbors.
Print all peak elements.

Input:
7

1 3 2 5 4 6 3

Output:
3 5 6

```python
N = int(input())
arr = list(map(int, input().split()))

peak_elements = []

for i in range(1, N):
    if (arr[i] > arr[(i - 1)]) and (arr[i] > arr[(i + 1)]):
        peak_elements.append(arr[i])

print(peak_elements if peak_elements else "no")
```

#### Question 20: A leader is an element greater than all elements to its right.
Print all leaders in order of appearance.

Input:
6

16 17 4 3 5 2

Output:
17 5 2

```python
N = int(input())
arr = list(map(int, input().split()))

L = []

for idx in range(N):
    if idx == N - 1:
        L.append(arr[idx])
    elif arr[idx] > max(arr[idx + 1 :]):
        L.append(arr[idx])
print(L)
```

```python
# alternative method
N = int(input())
arr = list(map(int, input().split()))

Leaders = []
max_from_right = float("-inf")

for i in range(N - 1, -1, -1):
    if arr[i] > max_from_right:
        Leaders.append(arr[i])
        max_from_right = arr[i]

Leaders.reverse()

print(Leaders)
```

#### Exercise 21: Array Equilibrium

Find index where sum of left elements = sum of right elements.
Print index (0-based). If none, print -1.

Input:
7

1 2 3 4 3 2 1

Output:
3

(1+2+3 = 3+2+1)

```python
N = int(input())
arr = list(map(int, input().split()))

L = []
for i in range(1, N):
    left_sum = sum(arr[i:])
    right_sum = sum(arr[: i + 1])

    if left_sum == right_sum:
        L.append(i)

print(L)
```

```python
N = int(input())
arr = list(map(int, input().split()))

total_sum = sum(arr)
left_sum = 0
results = []

for i in range(N):
    right_sum = total_sum - left_sum - arr[i]

    if left_sum == right_sum:
        results.append(i)
    left_sum += arr[i]

print(results)
```
