#### Exercise 1.3: String Analysis
```python
# Write a program that:
# 1. Counts vowels and consonants in a string
# 2. Finds the first occurrence of each character
# 3. Checks if two strings are anagrams
# 4. Reverses a string without slicing
```

````python
def string_analysis(text1):
    vowel_count = 0
    consonants_count = 0
    vowels = "aeiouAEIOU"

    for char in text1:
        if char in vowels:
            vowel_count += 1
        else:
            consonants_count += 1

    firstOccurrance = {}  # defining a dict to store the index of each char.
    for idx, char in enumerate(text1):
        if (
            char not in firstOccurrance
        ):  # if char is not present in dict, it means it is his first occurance,
            firstOccurrance[char] = idx

    reverse_text = ""
    for char in text1:
        reverse_text = char + reverse_text  # S[1] + S[0] and so onnn.

    return {
        "vowel_count": vowel_count,
        "consonants_count": consonants_count,
        "reverse_text": reverse_text,
        "first_occurrance": firstOccurrance,
    }


S = input()
string_analysis(S)
````

```python
# Write a program that classifies a number as:
# 1. Positive, Negative, or Zero
# 2. Even or Odd
# 3. Prime or Composite
# 4. Perfect Square or Not
# 5. Armstrong Number or Not (e.g., 153 = 1³ + 5³ + 3³)
```

````python
N = int(input())


# 1. Positive, Negative, or Zero
def int_check(num):
    if num == 0:
        arson = "Zero"
    elif num < 0:
        arson = "-Ve"
    else:
        arson = "+Ve"
    return arson


# 2. Even or Odd
def odd_even(num):
    if num % 2 == 0:
        arson = "Even"
    else:
        arson = "Odd"
    return arson


# 3. Prime or Composite
def chk_prime(num):
    if num <= 1:
        return "Not Prime"
    elif num == 2:
        return "Prime"
    elif num % 2 == 0:
        return "Not Prime"
    else:
        for ch in range(3, num, 2):
            if num % ch == 0:
                return "Not Prime"
        return "Prime"
        # return arson


# 4. Perfect Square or Not
def perfect_sq(num):
    root = num**0.5

    if root == int(root):
        arson = "Perfect sq"
    else:
        arson = "imperfect sq"
    return arson


# 5. Armstrong Number or Not (e.g., 153 = 1³ + 5³ + 3³)
def armstrong_num(num):
    value = []
    arson = "no"
    str_num = str(num)
    for i in str_num:
        chk = int(i) ** len(str_num)
        value.append(chk)
    if sum(value) == num:
        arson = "yes"
    return arson


print(int_check(N))
print(odd_even(N))
print(chk_prime(N))
print(perfect_sq(N))
print(armstrong_num(N))
````

````python
def is_prime(num: int) -> bool:
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    else:
        for i in range(3, int(num**0.5), 2):
            if num % i == 0:
                return False
        return True


is_prime(5)
````

````python
def gcd(a, b: int) -> int:
    """to find largest int that divides both numbers"""
    results_a = []
    results_b = []
    for i in range(1, a + 1):
        if a % i == 0:
            results_a.append(i)

    for i in range(1, b + 1):
        if b % i == 0:
            results_b.append(i)
    results = []
    for i in results_a:
        if i in results_b:
            results.append(i)
    return max(results)


gcd(6, 8)
````

````python
def lcm(a, b: int) -> int:
    results_a = []
    results_b = []

    for i in range(1, 100):
        a_v = a * i
        b_v = b * i
        results_a.append(a_v)
        results_b.append(b_v)

        for num in results_a:
            if num in results_b:
                return num


lcm(12, 18)
````

````python
# using while loop
def lcm(a: int, b: int) -> int:
    """small common int in multiples of both numbers"""
    i = 1
    while True:
        multiple = a * i
        if multiple % b == 0:
            return multiple
        i += 1


print(lcm(12, 18))
````

````python
def power(base: int, exp: int) -> int:
    """Calculate base^exp"""
    if base <= 0:
        return 0
    results = 1  # defining results value, (in case exp = 0)
    for _ in range(exp):  # to run our loop exp numbers of times
        results = results * base
    return results


power(2, 3)
````

````python
def sqrt_approx(n: int) -> int:
    """Approximate square root using binary search"""
    if n < 0:
        raise ValueError("Int is less than zero")
    if n == 0 or n == 1:
        return 1

    low, high = 1, n  # (sqrt must be between 1 and n)
    while low <= high:
        mid = (low + high) // 2
        sq = mid * mid

        if sq == n:
            return mid
        elif sq < n:
            low = mid + 1
        else:
            high = mid - 1
    return high


sqrt_approx(1)
````

````python
def factorial(n: int) -> int:
    """Calculate factorial recursively and iteratively"""
    value = 1
    for i in range(1, n + 1):
        value = value * i
    return value


factorial(4)
````

````python
def find_max(lst: list) -> int:
    """Return maximum element"""
    lst.sort(reverse=True)
    return lst[0]


arr = [2, 3, 1, 5]
find_max(arr)
````

````python
def find_min(lst: list) -> int:
    """Return minimum element"""
    lst.sort()
    return lst[0]


find_min(arr)
````

````python
def reverse_list(lst: list) -> list:
    """Return reversed list"""
    return lst[::-1]


arr = [1, 8, 7, 5]
reverse_list(arr)
````

````python
def remove_duplicates(lst: list) -> list:
    """Remove duplicates preserving order"""
    duplicates = []
    for i in lst:
        if i in duplicates:
            lst.remove(i)
        else:
            duplicates.append(i)
    return duplicates


arr = [2, 2, 2, 1, 4, 1]
remove_duplicates(arr)
````

````python
def merge_sorted_lists(list1: list, list2: list) -> list:
    """Merge two sorted lists"""
    results = []
    for i in list1:
        results.append(i)

    for i in list2:
        results.append(i)
    return results


lst = [1, 3, 5]
lst2 = [2, 4]
merge_sorted_lists(lst, lst2)
````

````python
# Problem 3: Composite Number Check
# Print 'yes' if composite, else 'no'
factors = []


def is_composite(num):
    for i in range(1, num + 1):
        if num % i == 0:
            factors.append(i)

    if len(factors) > 2:
        return True


N = int(input())
if is_composite(N):
    print("yes")
else:
    print("noo")
````

````python
# Find First and Last Position of a Target
# 🔹 Problem Statement
# Given a sorted array of integers (may contain duplicates) and a target value, find the starting and ending position of the target. If the target is not found, return [-1, -1].

arr = list(map(int, input().split()))
target = int(input())


def target_position(a: list, b: int) -> list:
    low, high = 0, len(a) - 1
    results = []
    while low <= high:
        mid = (low + high) // 2
        if a[mid] == b:
            results.append(mid)
        elif a[mid] < b:
            low = mid + 1
        else:
            high = mid - 1
    return results


print(target_position(arr, target))
````

````python
def first_find(a: list[int], b: int) -> int:  # args a -> list[int], b -> target number
    low, high = 0, len(a) - 1  #
    first = -1  # initial value

    while low <= high:
        mid = (low + high) // 2
        if a[mid] == b:  # once we got our element, we ll add its index to var(first)
            first = mid
            high = mid - 1  #
        elif a[mid] < b:
            low = mid + 1
        else:
            high = mid - 1
    return first


def find_last(a: list[int], b: int) -> int:
    low, high = 0, len(a) - 1
    last = -1

    while low <= high:
        mid = (low + high) // 2

        if a[mid] == b:
            last = mid
            low = mid + 1
        elif a[mid] < b:
            low = mid + 1
        else:
            high = mid - 1
    return last


def find_range(a: list[int], b: int) -> list:
    first = first_find(a, b)
    if first == -1:
        return (-1, -1)
    last = find_last(a, b)
    if last:
        return (first, last)


arr = [5, 7, 7, 8, 8, 8, 10]
target = 9
# print(first_find(arr,target))
# print(find_last(arr,target))
print(find_range(arr, target))
````

````python
# Exercise 2 Recap: Search in Rotated Sorted Array
# Goal: Find target in a rotated sorted array (e.g., [4,5,6,7,0,1,2]) in O(log n) time.


def rotated_array(a: list[int], b: int) -> int:
    low, high = 0, len(a) - 1
    while low <= high:
        mid = (low + high) // 2

        if a[mid] == b:
            return mid

        # check which side is sorted, left or right
        if a[low] < a[mid]:
            # check whether the target element is in left side
            if a[low] <= b < a[mid]:
                # if target in left side, search left otherwise right.
                high = mid - 1
            else:  # switching to right side
                low = mid + 1
        else:
            if (
                a[high] >= b > a[mid]
            ):  # checking target lies in right side,if yes, search right side
                low = mid + 1
            else:  # if target is not in right side, search in left side
                high = mid = 1
    return -1


arr = [4, 5, 6, 7, 0, 1, 2]
K = 7
print(rotated_array(arr, K))
````
